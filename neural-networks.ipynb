```python
from sklearn.datasets import make_moons
from sklearn.neural_network import MLPClassifier
import mglearn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X,y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=42)
```

```python
mlp = MLPClassifier(solver='lbfgs', random_state=0)
mlp.fit(X_train,y_train)
```

```python
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1],y_train)
plt.xlabel('f 0')
plt.ylabel('f 1')
```

```python
print("Accuracy on training set {:.8f}".format(mlp.score(X_train,y_train)))
print("Accuracy on test set {:.8f}".format(mlp.score(X_test,y_test)))
```

```python
mlp10 = MLPClassifier(solver='lbfgs', hidden_layer_sizes=[10], max_iter=1000, random_state=0)
mlp10.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp10, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1],y_train)
plt.xlabel('f 0')
plt.ylabel('f 1')
```

```python
print("Accuracy on training set {:.8f}".format(mlp10.score(X_train,y_train)))
print("Accuracy on test set {:.8f}".format(mlp10.score(X_test,y_test)))
```

```python
mlp10_10 = MLPClassifier(solver='lbfgs', hidden_layer_sizes=[10,10], max_iter=1000, random_state=0)
mlp10_10.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp10_10, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1],y_train)
plt.xlabel('f 0')
plt.ylabel('f 1')
```

```python
print("Accuracy on training set {:.8f}".format(mlp10_10.score(X_train,y_train)))
print("Accuracy on test set {:.8f}".format(mlp10_10.score(X_test,y_test)))
```

```python
mlp10_10_alpha = MLPClassifier(solver='lbfgs', hidden_layer_sizes=[10,10], max_iter=1000, alpha=.1,random_state=0)
mlp10_10_alpha.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp10_10_alpha, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1],y_train)
plt.xlabel('f 0')
plt.ylabel('f 1')
```

```python
print("Accuracy on training set {:.8f}".format(mlp10_10_alpha.score(X_train,y_train)))
print("Accuracy on test set {:.8f}".format(mlp10_10_alpha.score(X_test,y_test)))
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

# Split the cancer data set
X_cancer_train,X_cancer_test,y_cancer_train,y_cancer_test = train_test_split(cancer.data,cancer.target,random_state=0)

# Scale the data to have a mean of 0 and a variance of 1
train_mean = X_cancer_train.mean(axis=0)
std_train = X_cancer_train.std(axis=0)
X_cancer_train = (X_cancer_train - train_mean) / std_train
X_cancer_test = (X_cancer_test - train_mean) / std_train

# Train and test the model
mlp_cancer = MLPClassifier(max_iter=1000, alpha=.1, random_state=42)
mlp_cancer.fit(X_cancer_train,y_cancer_train)
print("Accuracy on training set {:.8f}".format(mlp_cancer.score(X_cancer_train,y_cancer_train)))
print("Accuracy on test set {:.8f}".format(mlp_cancer.score(X_cancer_test,y_cancer_test)))
```
